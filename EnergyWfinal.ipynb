```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

# Upload the CSV
from google.colab import files
uploaded = files.upload()

# Read the file
df = pd.read_csv('PJM_Load_hourly.csv', parse_dates=['Datetime'], index_col='Datetime')
df.rename(columns={'PJM_Load_MW': 'Energy'}, inplace=True)

# Plot the original energy usage
plt.figure(figsize=(12, 6))
plt.plot(df['Energy'], label='Hourly Energy Consumption (MW)')
plt.title('Energy Usage Over Time')
plt.xlabel('Time')
plt.ylabel('Energy Consumption (MW)')
plt.legend()
plt.show()

# Scale the data
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(df[['Energy']])

# Create dataset function
def create_dataset(data, time_step=24):
    X, y = [], []
    for i in range(len(data) - time_step):
        X.append(data[i:i + time_step])
        y.append(data[i + time_step])
    return np.array(X), np.array(y)

time_step = 24
X, y = create_dataset(scaled_data, time_step)
X = X.reshape(X.shape[0], X.shape[1], 1)

# Split into training and testing
train_size = int(len(X) * 0.8)
X_train, y_train = X[:train_size], y[:train_size]
X_test, y_test = X[train_size:], y[train_size:]

# Build LSTM model
model = Sequential()
model.add(LSTM(64, return_sequences=True, input_shape=(time_step, 1)))
model.add(LSTM(32))
model.add(Dense(1))

model.compile(optimizer='adam', loss='mean_squared_error')
model.fit(X_train, y_train, epochs=10, batch_size=32, verbose=1)

# Predict and unscale
predictions = model.predict(X_test)
predictions_unscaled = scaler.inverse_transform(predictions)
y_test_unscaled = scaler.inverse_transform(y_test.reshape(-1, 1))

# Plot actual vs predicted
plt.figure(figsize=(12, 6))
plt.plot(y_test_unscaled, label='Actual')
plt.plot(predictions_unscaled, label='Predicted')
plt.title('Actual vs Predicted Energy Usage')
plt.xlabel('Time')
plt.ylabel('Energy Consumption (MW)')
plt.legend()
plt.show()

# 30-day forecast
forecast_steps = 24 * 30
last_window = scaled_data[-time_step:]
forecast = []

for _ in range(forecast_steps):
    input_seq = last_window.reshape(1, time_step, 1)
    next_pred = model.predict(input_seq)[0][0]
    forecast.append(next_pred)
    last_window = np.append(last_window[1:], [[next_pred]], axis=0)

forecast = np.array(forecast).reshape(-1, 1)
forecast_unscaled = scaler.inverse_transform(forecast)
forecast_unscaled = np.clip(forecast_unscaled, 0, None)

# Create time index for forecast
recent_data = df['Energy'][-336:]
recent_index = recent_data.index
forecast_index = pd.date_range(start=recent_index[-1] + pd.Timedelta(hours=1), periods=forecast_steps, freq='H')

# Plot recent + forecast
plt.figure(figsize=(16, 6))
plt.plot(recent_index, recent_data.values, label='Recent Actual (2 weeks)', color='blue')
plt.plot(forecast_index, forecast_unscaled, label='Forecast (Next 30 Days)', color='orange')
plt.title('Energy Forecast for Next 30 Days')
plt.xlabel('Time')
plt.ylabel('Energy Consumption (MW)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Plot ONLY the 30-day forecast
plt.figure(figsize=(16, 6))
plt.plot(forecast_index, forecast_unscaled, color='orange', label='Forecast Only (30 Days)')
plt.title('🔮 Forecast: Energy Consumption (Next 30 Days)')
plt.xlabel('Time')
plt.ylabel('Energy Consumption (MW)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
